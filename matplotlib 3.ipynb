```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[51,34,67,89,54,32,54]
min_t=[23,34,43,21,45,33,44]
avg_t=[35,39,35,33,34,36,37]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(days,max_t,label="maxTemp")
plt.plot(days,min_t,label="minTemp")
plt.plot(days,avg_t,label="AvgTemp")

plt.legend()
```
