```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.bar(ypos,revenue,label="revenue")
plt.ylabel("revenue")
plt.title("US tech stocks")
plt.xticks(ypos,company)
plt.bar(ypos,profit,label="profit")
plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(ypos,revenue,width=0.4,label="revenue")
plt.ylabel("revenue")
plt.title("US tech stocks")
plt.xticks(xpos-0.2,company)
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.barh(ypos,revenue,label="revenue")
plt.ylabel("revenue")
plt.title("US tech stocks")


plt.yticks(xpos-0.2,company)
plt.barh(xpos+0.2,profit,label="profit")
plt.legend()
```
