```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["home rent","food","phone/internet bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.0f%%',shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.0f%%',shadow=True,explode=[0,0.2,0,0,0])
plt.show()
```
